```python
import pandas as pd
import chardet

path = "/content/drive/MyDrive/Colab Notebooks/train_LSTM.csv"

# Detect the file's encoding
with open(path, 'rb') as f:
    result = chardet.detect(f.read())
encoding = result['encoding']

# Read the CSV file using the detected encoding
df = pd.read_csv(path, encoding=encoding, on_bad_lines='warn')

df.head()
```

```python
df = df.dropna()
```

```python
X = df.drop('label', axis=1)
y = df['label']
```

```python
X.shape
```

```python
y.shape
```

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
import nltk
import re
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.stem import PorterStemmer
```

```python
messages = X.copy()
messages.reset_index(inplace=True)
messages.head()
```

```python
## Data Preprocessing
corpus = []
for i in range(0, len(messages)):
  sent = re.sub('[^a-zA-Z]', ' ',messages['text'][i])
  sent = sent.lower()
  sent = sent.split()
  ps = PorterStemmer()
  sent = [ps.stem(word)for word in sent if word not in nltk.corpus.stopwords.words('english')]
  sent = ' '.join(sent)
  corpus.append(sent)
```

```python
print(corpus)
```

```python
# One Hot Encoding
voc_size = 5000
onehot_repr = [one_hot(words, voc_size) for words in corpus]
onehot_repr

```

```python
# Padding
max_len = 30
embedded_docs = pad_sequences(onehot_repr, padding = 'post', maxlen = max_len)
print(embedded_docs)
```

```python
print(len(embedded_docs[0]))
print(len(embedded_docs[7]))
```

```python
import numpy as np
X_final= np.array(embedded_docs)
y_final = np.array(y)
```

```python
X_final.shape, y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
# Create Model
in_dim = voc_size
out_dim = 40
model = Sequential()
model.add(Embedding(in_dim, out_dim, input_length = max_len))
model.add(LSTM(100))
model.add(Dense(1, activation = 'sigmoid'))
model.summary()
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
model.fit(X_train, y_train, validation_data = (X_test, y_test), epochs = 10, batch_size = 64)
```

```python
y_pred = model.predict(X_test)
```

```python
error = model.evaluate(X_test, y_test)
```

```python

accuracy = error[1]
print("Accuracy: {:.2f}%".format(accuracy * 100))
```
